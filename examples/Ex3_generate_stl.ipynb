# Example 3 - Generating an stl file from a wing geometry

### Define wing geometry

```python
import AXIS_wing_definitions as AX_wings
# Define front wing
wing = AX_wings.bsc_810(1e6, afoil='naca1710', nsegs=40, plot_flag=False)
wing.plot2D()
```

### Export to stl

```python
%matplotlib widget

# recommended scaling dimensions by 1000, i.e. from m to mm, as most CAD softwares use mm as default.
scale_factor = 1000
# pitch mounting angle in degrees, useful if the wing will be mounted at an angle relative to the origin axes
mounting_angle = 1
# Recommend high resolution when exporting for CAD processing/3D printing
stl_resolution = 'high' # 'low', 'medium', or 'high'

file_name = 'BSC810_high_res'
wing.export_wing_2_stl(file_name,
                        SF=scale_factor,  
                        mounting_angle=mounting_angle,
                        resolution=stl_resolution,
                        plot_flag=True)
```
